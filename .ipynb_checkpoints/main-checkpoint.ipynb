```python
import torch
```

```python
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

from PIL import Image
import matplotlib.pyplot as plt

import torchvision.transforms as transforms
import torchvision.models as models

import copy
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
device
```

```python
IMGSIZE = 512 if torch.cuda.is_available() else 128

loader = transforms.Compose([
    transforms.Resize(IMGSIZE),
    transforms.ToTensor()
])
```

```python
def image_loader(image_name):
    image = Image.open(image_name)
    image = loader(image).unsqueeze(0)
    return image.to(device, torch.float)
```

```python
content_image = image_loader("content.jpg")
style_image = image_loader("style.jpg")
```

```python
assert style_image.size() == content_image.size()
```

```python
unloader = transforms.ToPILImage()

plt.ion()

def imshow(tensor, title = None):
    image = tensor.cpu().clone()
    image = image.squeeze(0)
    image = unloader(image)
    if title:
        plt.title(title)
    plt.pause(0.001)

plt.figure()
imshow(style_image, title = "Style")
plt.figure()
imshow(content_image, title = "Content")
```
